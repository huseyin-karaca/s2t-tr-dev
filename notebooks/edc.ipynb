```python
from datasets import load_from_disk
import numpy as np

# 1. Veri setini diskten yükle
veri_yolu = "../data/processed/voxpopuli_asr_features_poc"
dataset = load_from_disk(veri_yolu)

# --- GENEL BAKIŞ ---
print("--- VERİ SETİ ÖZETİ ---")
print(dataset)
print(f"Toplam Örnek Sayısı: {len(dataset)}")
print(f"Sütunlar: {dataset.column_names}\n")

# --- TEK BİR ÖRNEĞİ İNCELEME ---
print("--- İLK ÖRNEĞİN DETAYLARI ---")
ornek = dataset[0] # İlk satırı alıyoruz

# 1. Metin Karşılaştırması
# Voxpopuli veri setinde orijinal metin genelde "normalized_text" veya "raw_text" sütununda bulunur
orijinal_metin = ornek.get("normalized_text", "Bulunamadı") 
print(f"Gerçek Metin (Ground Truth) : {orijinal_metin}")
print(f"Wav2Vec2 Tahmini          : {ornek['w2v2_transcription']}\n")

# 2. Encoder Matrisinin İncelenmesi
# Hatırlarsan kaydederken .tolist() ile düz Python listesine çevirmiştik.
# Makine öğrenmesi işlemlerine devam etmek için bunu tekrar Numpy veya PyTorch tensörüne çeviririz.
encoder_matrisi = np.array(ornek["w2v2_encoder_state"])

print("--- ENCODER STATE (ÖZNEKLİK) DETAYLARI ---")
print(f"Veri Tipi: {type(encoder_matrisi)}")
print(f"Matris Boyutu (Shape): {encoder_matrisi.shape}")

# Boyutların Anlamı:
zaman_adimi = encoder_matrisi.shape[0]
vektor_boyutu = encoder_matrisi.shape[1]
print(f"-> Bu ses dosyası {zaman_adimi} zaman adımına (frame) bölünmüş.")
print(f"-> Her bir zaman adımı {vektor_boyutu} boyutlu bir vektörle (hidden state) temsil ediliyor.")
```

```python

# --- TEK BİR ÖRNEĞİ İNCELEME ---
print("--- İKİNCİ ÖRNEĞİN DETAYLARI ---")
ornek = dataset[1] # İkinci satırı alıyoruz

# 1. Metin Karşılaştırması
# Voxpopuli veri setinde orijinal metin genelde "normalized_text" veya "raw_text" sütununda bulunur
orijinal_metin = ornek.get("normalized_text", "Bulunamadı") 
print(f"Gerçek Metin (Ground Truth) : {orijinal_metin}")
print(f"Wav2Vec2 Tahmini          : {ornek['w2v2_transcription']}\n")

# 2. Encoder Matrisinin İncelenmesi
# Hatırlarsan kaydederken .tolist() ile düz Python listesine çevirmiştik.
# Makine öğrenmesi işlemlerine devam etmek için bunu tekrar Numpy veya PyTorch tensörüne çeviririz.
encoder_matrisi = np.array(ornek["w2v2_encoder_state"])

print("--- ENCODER STATE (ÖZNEKLİK) DETAYLARI ---")
print(f"Veri Tipi: {type(encoder_matrisi)}")
print(f"Matris Boyutu (Shape): {encoder_matrisi.shape}")

# Boyutların Anlamı:
zaman_adimi = encoder_matrisi.shape[0]
vektor_boyutu = encoder_matrisi.shape[1]
print(f"-> Bu ses dosyası {zaman_adimi} zaman adımına (frame) bölünmüş.")
print(f"-> Her bir zaman adımı {vektor_boyutu} boyutlu bir vektörle (hidden state) temsil ediliyor.")
```
