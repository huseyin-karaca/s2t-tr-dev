```python
from dataclasses import dataclass

@dataclass
class Dataset:
    name: str
    subset: str
    split: dict

    def load(self, cache_dir="../data/raw"):
        return load_dataset(
            self.name,
            revision="refs/convert/parquet",
            data_files={self.split: f"{self.subset}/{self.split}/*.parquet"},
            cache_dir=cache_dir,
        )

ds1 = Dataset(
    name="edinburghcstr/ami",
    subset="ihm",
    split="test")


loaded_dataset1 = load_dataset(
    ds1.name,
    revision="refs/convert/parquet",
    data_files={ds1.split: f"{ds1.subset}/{ds1.split}/*.parquet"},
    cache_dir="../data/raw",
)


ds2 = Dataset(
    name="openslr/librispeech_asr",
    subset="other",
    split="test")


loaded_dataset2 = load_dataset(
    ds2.name,
    revision="refs/convert/parquet",
    data_files={ds2.split: f"{ds2.subset}/{ds2.split}/*.parquet"},
    cache_dir="../data/raw",
)
```
