```python
!pip install google-api-python-client google-auth-httplib2 google-auth-oauthlib -q

from google.oauth2 import service_account
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
import os
import json
from google.colab import userdata

# 1. Json dosyanızın yolu
SERVICE_ACCOUNT_FILE = json.loads(userdata.get('GCP_CREDENTIALS'))
SCOPES = ['https://www.googleapis.com/auth/drive']

# 2. Sessiz Kimlik DoĐrulama (UI Prompt ÇIKMAZ)
creds = service_account.Credentials.from_service_account_info(
        SERVICE_ACCOUNT_FILE, scopes=SCOPES)

# 3. Drive servisini başlat
drive_service = build('drive', 'v3', credentials=creds)

def upload_to_drive_silently(local_file_path, drive_folder_id):
    """Lokaldeki bir dosyayı UI onayı olmadan Drive'a yükler."""

    if not os.path.exists(local_file_path):
        print(f"❌ Dosya bulunamadı: {local_file_path}")
        return

    file_metadata = {
        'name': os.path.basename(local_file_path),
        'parents': [drive_folder_id]
    }

    # Parquet veya herhangi bir binary dosya için mimetype
    media = MediaFileUpload(local_file_path, mimetype='application/octet-stream', resumable=True)

    print(f"🚀 {os.path.basename(local_file_path)} Drive'a yükleniyor...")

    file = drive_service.files().create(body=file_metadata, media_body=media, fields='id').execute()

    print(f"✅ Başarelı! Dosya ID: {file.get('id')}")

# --- KULLANIM ---
DRIVE_KLASOR_ID = "1aPzmaqt5wqu95X0Vixnfmmwx9gXTGwDX" # Adım 2'de URL'den aldığınız ID
LOCAL_PARQUET = "data/processed/edinburghcstr_ami/combined_features.parquet"

upload_to_drive_silently(LOCAL_PARQUET, DRIVE_KLASOR_ID)
```

```python
import pyarrow.parquet as pq
import numpy as np

t = pq.read_table("data/processed/edinburghcstr_ami/combined_features-w2v2-large.parquet")
wers = np.stack([
    t["hubert_wer"].to_numpy(),
    t["whisper_wer"].to_numpy(),
    t["w2v2_wer"].to_numpy(),
], axis=1)  # (N, 3)

names = ["hubert", "whisper", "wav2vec2"]
oracle_winner = wers.argmin(axis=1)
print("Oracle win-rate:")
for i, n in enumerate(names):
    print(f"  {n:10s}: {(oracle_winner == i).mean():.3f}")
print(f"\nMean WER per model:")
for i, n in enumerate(names):
    print(f"  {n:10s}: {wers[:, i].mean():.3f}")
```

```python
import logging
from datasets import load_dataset, concatenate_datasets

logger = logging.getLogger(__name__)

def create_combined_parquet(
    hubert_path: str,
    whisper_path: str,
    w2v2_path: str,
    output_path: str
):
    logger.info("Interim dosyalar memory-mapped olarak açılıyor...")

    ds_hubert = load_dataset("parquet", data_files=hubert_path, split="train")
    ds_whisper = load_dataset("parquet", data_files=whisper_path, split="train")
    ds_w2v2 = load_dataset("parquet", data_files=w2v2_path, split="train")

    assert len(ds_hubert) == len(ds_whisper) == len(ds_w2v2), "Veri seti uzunlukları eşleşmiyor!"

    # ÇAKIŞMA ÇÖZÜMÜ:
    # Hubert datasetindeki sütun isimlerini referans alıyoruz
    base_columns = set(ds_hubert.column_names)

    # Whisper ve W2V2'deki ortak sütunları (ground_truth vb.) bulup siliyoruz
    whisper_duplicates = list(base_columns.intersection(ds_whisper.column_names))
    ds_whisper = ds_whisper.remove_columns(whisper_duplicates)

    w2v2_duplicates = list(base_columns.intersection(ds_w2v2.column_names))
    ds_w2v2 = ds_w2v2.remove_columns(w2v2_duplicates)

    logger.info(f"Çakışan sütunlar silindi. Whisper'dan silinenler: {whisper_duplicates}, W2V2'den silinenler: {w2v2_duplicates}")

    logger.info("Sütunlar sanal olarak birleştiriliyor...")
    combined_ds = concatenate_datasets([ds_hubert, ds_whisper, ds_w2v2], axis=1)

    logger.info(f"Tekil dosya yazılıyor: {output_path} (Bu işlem RAM'i şişirmeden batch'ler halinde yapılır)")
    combined_ds.to_parquet(output_path)
    logger.info("İşlem tamamlandı!")

# Kullanımı:
create_combined_parquet(
    "/content/s2t-tr-dev/data/interim/facebook_voxpopuli/facebook_hubert-large-ls960-ft.parquet",
    "/content/s2t-tr-dev/data/interim/facebook_voxpopuli/openai_whisper-base.parquet",
    "/content/s2t-tr-dev/data/interim/facebook_voxpopuli/facebook_wav2vec2-large-robust-ft-swbd-300h.parquet",
    "/content/s2t-tr-dev/data/processed/facebook_voxpopuli/combined_features.parquet")
```
